```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns

import os
for dirname, _, filenames in os.walk('./'):
    for filename in filenames:
        print(os.path.join(dirname, filename))


train_data=pd.read_csv("./train.csv")
test_data=pd.read_csv("./test.csv")
```

```python
# Parsing data

# MSZoning

d = {'A': 0, 'C (all)': 1, 'FV': 2, 'I' : 3,'RH': 4,'RL': 5, 'RP': 6,'RM': 7}
train_data['MSZoning'] = train_data['MSZoning'].map(d)
test_data['MSZoning'] = test_data['MSZoning'].map(d)

```

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["diagnosis"]

features = test_data.columns

X = pd.get_dummies(train_data[features])
print(X)
X_test = pd.get_dummies(test_data[features])
print(X_test)
model = RandomForestClassifier(n_estimators=200, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'id': test_data.id, 'diagnosis': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```

```python
# Grading submission

predictions=pd.read_csv("./submission.csv")

test_data=pd.read_csv("./test80.csv")

total=0
correct=0

for i in range(len(predictions)):
    predictedRow=predictions.iloc[i]
    predictedId=predictedRow["id"]
    predictedDiagnosis=predictedRow["diagnosis"]

    testRow=test_data.iloc[i]
    testId=testRow["id"]
    testDiagnosis=testRow["diagnosis"]

    if testId!=predictedId:
        print("not same id at row "+str(i))
    else:
        total=total+1
        if predictedDiagnosis==testDiagnosis:
            correct=correct+1

print("Total rows: "+str(total)+"\nCorrect diagnosis: "+str(correct)+"\nPercentage correct: "+str(correct/total))



```
