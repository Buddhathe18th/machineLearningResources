```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns

import os
for dirname, _, filenames in os.walk('./'):
    for filename in filenames:
        print(os.path.join(dirname, filename))


train_data=pd.read_csv("./train.csv")
test_data=pd.read_csv("./test.csv")
```

```python
# Parsing data

# MSZoning

d = {'A': 0, 'C (all)': 1, 'FV': 2, 'I' : 3,'RH': 4,'RL': 5, 'RP': 6,'RM': 7}
train_data['MSZoning'] = train_data['MSZoning'].map(d)
test_data['MSZoning'] = test_data['MSZoning'].map(d)

# LotFrontage

train_data.loc[train_data['LotFrontage']=='NA','LotFrontage'] = 0
test_data.loc[test_data['LotFrontage']=='NA','LotFrontage'] = 0

# Street

d = {'Pave': 0, 'Grvl': 1}
train_data['Street'] = train_data['Street'].map(d)
test_data['Street'] = test_data['Street'].map(d)

# Alley

d = {'NA': 0, 'Pave': 1, 'Grvl': 2}
train_data['Alley'] = train_data['Alley'].map(d)
test_data['Alley'] = test_data['Alley'].map(d)

# LotShape

d = {'Reg': 0, 'IR1': 1, 'IR1': 2, 'IR1': 3}
train_data['LotShape'] = train_data['LotShape'].map(d)
test_data['LotShape'] = test_data['LotShape'].map(d)
```

```python
from sklearn.ensemble import RandomForestRegressor

y = train_data["SalePrice"]

features = test_data.columns

X_all = pd.get_dummies(train_data[features])
X_test=pd.get_dummies(test_data[features])
X = X_all[X_test.columns]

print(X.columns)
# X_test_all = pd.get_dummies(test_data[features])
# X_test=X_test_all[X.columns]
model = RandomForestRegressor(n_estimators=200, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'Id': test_data.Id, 'SalePrice': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```

```python
# Grading submission

predictions=pd.read_csv("./submission.csv")

test_data=pd.read_csv("./test80.csv")

total=0
correct=0

for i in range(len(predictions)):
    predictedRow=predictions.iloc[i]
    predictedId=predictedRow["id"]
    predictedDiagnosis=predictedRow["diagnosis"]

    testRow=test_data.iloc[i]
    testId=testRow["id"]
    testDiagnosis=testRow["diagnosis"]

    if testId!=predictedId:
        print("not same id at row "+str(i))
    else:
        total=total+1
        if predictedDiagnosis==testDiagnosis:
            correct=correct+1

print("Total rows: "+str(total)+"\nCorrect diagnosis: "+str(correct)+"\nPercentage correct: "+str(correct/total))



```
