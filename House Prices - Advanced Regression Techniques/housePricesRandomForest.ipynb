```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns

import os
for dirname, _, filenames in os.walk('./'):
    for filename in filenames:
        print(os.path.join(dirname, filename))


train_data=pd.read_csv("./train.csv")
test_data=pd.read_csv("./test.csv")
```

```python
# Parsing data

# MSZoning

d = {'A': 0, 'C (all)': 1, 'FV': 2, 'I' : 3,'RH': 4,'RL': 5, 'RP': 6,'RM': 7}
train_data['MSZoning'] = train_data['MSZoning'].map(d)
test_data['MSZoning'] = test_data['MSZoning'].map(d)

# LotFrontage

train_data.loc[train_data['LotFrontage']=='NA','LotFrontage'] = 0
test_data.loc[test_data['LotFrontage']=='NA','LotFrontage'] = 0

# Street

d = {'Pave': 0, 'Grvl': 1}
train_data['Street'] = train_data['Street'].map(d)
test_data['Street'] = test_data['Street'].map(d)

# Alley

d = {'NA': 0, 'Pave': 1, 'Grvl': 2}
train_data['Alley'] = train_data['Alley'].map(d)
test_data['Alley'] = test_data['Alley'].map(d)

# LotShape

d = {'Reg': 0, 'IR1': 1, 'IR1': 2, 'IR1': 3}
train_data['LotShape'] = train_data['LotShape'].map(d)
test_data['LotShape'] = test_data['LotShape'].map(d)

# Condition 2

d = {'Norm': 0, 'Feedr': 1, 'Artery': 2, 'RRNn': 3,"PosN": 4}
train_data['Condition2'] = train_data['Condition2'].map(d)
test_data['Condition2'] = test_data['Condition2'].map(d)

# Electrical

d = {'SBrkr': 0, 'FuseA': 1, 'FuseF': 2, 'FuseP': 3,"Mix": 4}
train_data['Electrical'] = train_data['Electrical'].map(d)
test_data['Electrical'] = test_data['Electrical'].map(d)

# Exterior1st

d = {'VinylSd': 0, 'HdBoard': 1, 'MetalSd': 2, 'Wd Sdng': 3,"Plywood": 4}
train_data['Exterior1st'] = train_data['Exterior1st'].map(d)
test_data['Exterior1st'] = test_data['Exterior1st'].map(d)

# Exterior2nd

d = {'VinylSd': 0, 'HdBoard': 1, 'MetalSd': 2, 'Wd Sdng': 3,"Plywood": 4}
train_data['Exterior2nd'] = train_data['Exterior2nd'].map(d)
test_data['Exterior2nd'] = test_data['Exterior2nd'].map(d)

# Heating

d = {'GasA': 0, 'GasW': 1, 'Grav': 2, 'Wall': 3,"OthW": 4}
train_data['Heating'] = train_data['Heating'].map(d)
test_data['Heating'] = test_data['Heating'].map(d)

# HouseStyle

d = {'1Story': 0, '2Story': 1, '1.5Fin': 2, 'SLvl': 3,"SFoyer": 4}
train_data['HouseStyle'] = train_data['HouseStyle'].map(d)
test_data['HouseStyle'] = test_data['HouseStyle'].map(d)

# GarageQual

d = {'TA': 0, 'NA': 1, 'Fa': 2, 'Gd': 3,"Ex": 4}
train_data['GarageQual'] = train_data['GarageQual'].map(d)
test_data['GarageQual'] = test_data['GarageQual'].map(d)

# MiscFeature

d = {'NA': 0, 'Shed': 1, 'Gar2': 2, 'Othr': 3,"TenC": 4}
train_data['MiscFeature'] = train_data['MiscFeature'].map(d)
test_data['MiscFeature'] = test_data['MiscFeature'].map(d)

# RoofMatl

d = {'CompShg': 0, 'Tar&Grv': 1, 'WdShngl': 2, 'WdShake': 3,"Metal": 4}
train_data['RoofMatl'] = train_data['RoofMatl'].map(d)
test_data['RoofMatl'] = test_data['RoofMatl'].map(d)

# PoolQC

d = {'NA': 1, 'Fa': 2, 'Gd': 3,"Ex": 4}
train_data['PoolQC'] = train_data['PoolQC'].map(d)
test_data['PoolQC'] = test_data['PoolQC'].map(d)

# Utilities

d = {'AllPub': 0, 'NoSeWa': 1}
train_data['Utilities'] = train_data['Utilities'].map(d)
test_data['Utilities'] = test_data['Utilities'].map(d)
```

```python
from sklearn.ensemble import RandomForestRegressor

y = train_data["SalePrice"]

features = test_data.columns

X_all = pd.get_dummies(train_data[features])
X_test=pd.get_dummies(test_data[features])
X=X_all
# X = X_all[X_test.columns]

print(X.columns)
# X_test_all = pd.get_dummies(test_data[features])
# X_test=X_test_all[X.columns]
model = RandomForestRegressor(n_estimators=200, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'Id': test_data.Id, 'SalePrice': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```

```python
# Grading submission

predictions=pd.read_csv("./submission.csv")

test_data=pd.read_csv("./test80.csv")

total=0
correct=0

for i in range(len(predictions)):
    predictedRow=predictions.iloc[i]
    predictedId=predictedRow["id"]
    predictedDiagnosis=predictedRow["diagnosis"]

    testRow=test_data.iloc[i]
    testId=testRow["id"]
    testDiagnosis=testRow["diagnosis"]

    if testId!=predictedId:
        print("not same id at row "+str(i))
    else:
        total=total+1
        if predictedDiagnosis==testDiagnosis:
            correct=correct+1

print("Total rows: "+str(total)+"\nCorrect diagnosis: "+str(correct)+"\nPercentage correct: "+str(correct/total))



```
